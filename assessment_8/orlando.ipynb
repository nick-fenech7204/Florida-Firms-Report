```python
df_cleaned = (df
    .replace({'SEX':{'001': 'Total'}})
    .replace({'SEX':{'002': 'Female'}})
    .replace({'SEX':{'003': 'Male'}})
    .replace({'SEX':{'004': 'Equally male/female'}})
    .replace({'SEX':{'096': 'Classified'}})
    .replace({'SEX':{'098': 'Unclassificable'}})
    .replace({'ETH_GROUP':{'001': 'Total'}})
    .replace({'ETH_GROUP':{'020': 'Hispanic'}})
    .replace({'ETH_GROUP':{'028': 'Equally Hispanic/non-Hispanic'}})
    .replace({'ETH_GROUP':{'029': 'Non-Hispanic'}})
    .replace({'ETH_GROUP':{'096': 'Classifiable'}})
    .replace({'ETH_GROUP':{'098': 'Unclassifiable'}})
    .replace({'RCPSZFI':{'511': 'Less than $5,000'}})
    .replace({'RCPSZFI':{'518': '$5,000-$9,999'}})
    .replace({'RCPSZFI':{'519': '$10,000-$24,999'}})
    .replace({'RCPSZFI':{'521': '$25,000-$49,999'}})
    .replace({'RCPSZFI':{'522': '$50,000-$99,999'}})
    .replace({'RCPSZFI':{'523': '$100,000-$249,999'}})
    .replace({'RCPSZFI':{'525': '$250,000-$499,999'}})
    .replace({'RCPSZFI':{'531': '$500,000-$999,999'}})
    .replace({'RCPSZFI':{'532': '$1,000,000 or more'}})
)

df_cleaned.head()
```

```python
df_cleaned = df_cleaned.drop(columns=['NAICS2017', 'NAICS2017_LABEL', 'EMPSZFI', 'metropolitan statistical area/micropolitan statistical area'])
```
